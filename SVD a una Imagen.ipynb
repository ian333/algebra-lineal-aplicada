```
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```
plt.style.use('classic')
imagen= Image.open("../Imagenes/imagen_ejemplo_frida_bredesen.jpg")
plt.imshow(imagen)
```

```
imagen_gr= imagen.convert('LA')
plt.imshow(imagen_gr)
```

```
imagen_mat=np.array(list(imagen_gr.getdata(band=0)),float)
print(imagen_mat)
```

```
imagen_mat.shape=(imagen_gr.size[1],imagen_gr.size[0])
print(imagen_mat)
```

```
print(imagen_mat.shape)
```

```
plt.imshow(imagen_mat,cmap='gray')
```

```
imagen_mat_2=imagen_mat/10
print(imagen_mat_2)
```

```
plt.imshow(imagen_mat_2)
```

```
print(np.max(imagen_mat_2))
print(np.max(imagen_mat))
```

```
U,D,V= np.linalg.svd(imagen_mat)
```

```
print(imagen_mat.shape)
print(U.shape)
print(D.shape)
print(V.shape)
```

## ¿Cuantos valores singulares se tendran que utilizar para que la imagen sea reconocible?

```
imagen_reconstruida=np.matrix(U[:,:1])*np.diag(D[:1])*np.matrix(V[:1,:])
plt.imshow(imagen_reconstruida,cmap='gray')
```

```

fig,ax = plt.subplots(4,2,figsize=(30,30))
for a in range(4):
    for b in range(2):
        i=(2**(a*2))*2**b
        imagen_reconstruida=np.matrix(U[:,:i])*np.diag(D[:i])*np.matrix(V[:i,:])
        ax[a,b].imshow(imagen_reconstruida,cmap='gray')
        ax[a,b].set_title(f'Imagen con {i} valores singulares')

plt.show()
```
