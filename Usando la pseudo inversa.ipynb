* Se quiere hallar la **solucion** del sistema de ecuaciones $Ax = b$, tal que x hace que $||Ax-b||_2$ sea **minima**.
* En principio el sistema de ecuaciones es **sobredeterminado**, lo que implica que las tres rectas **no van a cruzarse en un mismo punto**.
* La solucion dada a trav√©s de la pseudo inversa es tal que obedece a los **pesos de las ecuaciones**. Cada ecuacion como tal ejerce un efecto de gravedad que mueve el punto hacia ella.

```
 %matplotlib inline
 import numpy as np
 import matplotlib.pyplot as plt

 
```

```
x= np.linspace(-5,5,1000)

y_1= 1*x + 4
y_2= 2*x + 5
y_3= -3*x + 6

plt.plot(x,y_1)
plt.plot(x,y_2)
plt.plot(x,y_3)

plt.xlim(-2,2.5)
plt.ylim(-6,6)
plt.show()
```

```
matriz = np.array([[-1,1],[-2,1],[3,1]])
```

```
matriz_pse=np.linalg.pinv(matriz)

print(matriz_pse)
```

```
b = np.array([[4],[5],[6]])
b
```

```
resultado= matriz_pse.dot(b)
resultado
```

```
plt.plot(x,y_1)
plt.plot(x,y_2)
plt.plot(x,y_3)
plt.xlim(-2,2.5)
plt.ylim(-6,6)
plt.scatter(resultado[0],resultado[1])
plt.show()
```
