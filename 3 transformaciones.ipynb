```
A=np.array([[3,7],[5,2]])
print(A)
```

```
print('Circulo unitario')
graficanteMatriz(np.array([[1,0],[0,1]]))
plt.xlim(-2,2)
plt.ylim(-2,2)
plt.show()

print('Circulo unitario transformado')
graficanteMatriz(A)
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```
U,D,V = np.linalg.svd(A)
print(U)
```

```
print('Circulo unitario')
graficanteMatriz(np.array([[1,0],[0,1]]))
plt.xlim(-2,2)
plt.ylim(-2,2)
plt.show()

print('Primer Rotacion V')
graficanteMatriz(V)
plt.xlim(-3,3)
plt.ylim(-3,3)
plt.show()

print('Escala D')
graficanteMatriz(np.diag(D).dot(V))
plt.xlim(-9,9)
plt.ylim(-9,9)
plt.show()

print('Segunda Rotacion U')
graficanteMatriz(U.dot(np.diag(D).dot(V)))
plt.xlim(-9,9)
plt.ylim(-9,9)
plt.show()
```
